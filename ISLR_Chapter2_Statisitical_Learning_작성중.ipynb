$input$ $variable$: advertising budgets, denoted by $X$
\begin{align*}
 X_{1} &= \text{TV budget}\\
 X_{2} &= \text{Radio budget}\\
 X_{3} &= \text{newspaper budget}
\end{align*}

$output$ $variable$: sales, denoted by $Y$

\
Input go by different names, such as $predictors$, $independent$ $variables$, $features$, $variables$.
\
Output variable often called the $response$ or $dependent$ $variable$.
\
\
More generally, suppose that $Y \in \mathbb{R}$: quantitative response, $X = (X_{1}, X_{2}, \dots, X_{p})$: p differents fetures, we assume that there is some relationship between $Y$ and $X$, which can be written as $$ Y = f(X) + \epsilon.$$
\
$f$ is some fixed but unknowns funtion of $X_{1}, \dots, X_{p}$, and $\epsilon$ is a random error term, which is independent of $X$, $\mathbb{E}[\epsilon] = 0.$
In this formulation, $f$ represents the $systematic$ information that $X$ provides about $Y$.
\
\
In essence, statistical learning refers to a set of approaches for estimating $f$.

#### 2.1.1. Why Estimate $f$?
We may wish to estimate $f$: $prediction$ and $inference.$
\
\
In many situation, $X$ are readily available, $Y$ cannot be easily obtained.In this case, assume that error term averages to zero, we can predict $Y$ using $$ \hat{Y} = \hat{f}(X),$$ where $\hat{f}$: estimate for $f,$ and $\hat{Y}$: the resulting prediction for $Y$. 
\
In this setting, $\hat{f}$ is often treated as a $black$ $box.$


> $Reducible$ $error$ & $Inrreducible$ $error$

In general, $\hat{f}$ will not be a perfect estimate for $f$, and this inaccuracy will introduce some error. 
\
\
This error is $reducible$ because we can potentially improve the accuracy of $\hat{f}$ by using the most appropriate statistical learning technique to $f$.
\
\
If we find a perfect estimate for $f$, so that our estimated response took the form $\hat{Y} = f(X)$, but our prediction would still have some error $\epsilon$ by definion of $Y$. This is known as the $irreducible$ error, because no matter how well we estimate $f$, we cannot reduce the error introduced by $\epsilon$.
\
\
Consider a given estimate $\hat{f}$ and a set of features $X$, which yield the prediction $\hat{Y} = \hat{f}(X)$. Assume that for a moment that both $\hat{f}$ and X are fixed so that the only variability comes from $\epsilon$. Then, 
\begin{align*}
    \mathbb{E}[(Y-\hat{Y})^{2}] &= \mathbb{E}[(f(X) + \epsilon - \hat{f}(X))^{2}]\\
    &= \mathbb{E}[f^{2}(X) + \epsilon^{2} + \hat{f}^{2}(X) +2\epsilon f(X) - 2\epsilon \hat{f}(X) -2f(X)\hat{f}(X)] \\
    &= \mathbb{E}[f^{2}(X)] + \mathbb{E}[\epsilon^{2}] + \mathbb{E}[\hat{f}^{2}(X)] + 2\mathbb{E}[\epsilon f(X)] - 2\mathbb{E}[\epsilon \hat{f}(X)] - 2\mathbb{E}[f(X)\hat{f}(X)] \qquad \because linearity\\ 
    &= \mathbb{E}[f^{2}(X)] + \mathbb{E}[\epsilon^{2}] + \mathbb{E}[\hat{f}^{2}(X)] + 2\mathbb{E}[\epsilon]\mathbb{E}[f(X)] - 2\mathbb{E}[\epsilon]\mathbb{E}[\hat{f}(X)] - 2\mathbb{E}[f(X)\hat{f}(X)] \qquad \because independent\\
    &= \mathbb{E}[f^{2}(X)] + \mathbb{E}[\epsilon^{2}] + \mathbb{E}[\hat{f}^{2}(X)] - 2\mathbb{E}[f(X)\hat{f}(X)] \qquad \because \mathbb{E}[\epsilon] = 0\\
    &= f^{2}(X)+ \hat{f}^{2}(X) - 2f(X)\hat{f}(X) + \mathbb{E}(\epsilon^{2})  \qquad \because \hat{f} \text{ & } X: fixed \\
    &= \underbrace{[f(X) - \hat{f}(X)]^{2}}_{\text{Reducible}} + \underbrace{Var(\epsilon)}_{\text{Irreducible}}                    
\end{align*}
